```python
#Method Overriding

#Task 1: Define the Base Class Animal
class Animal:
  def speak(self):
    return "Animal speaks"
class Lion(Animal):
  def speak(self):
    return "Roar!"
class Cat(Animal):
  def speak(self):
    return "Meow!"
def make_animal_speak(animal):
  print(animal.speak())

if __name__ == "__main__":
    lion = Lion()
    cat = Cat()

    make_animal_speak(lion)
    make_animal_speak(cat)
```
