```python
#Method Overloading (Simulated)

#Task 1: Define the Printer Class with a Single Method
class Printer:
    def print_message(self, message, times=1):
        for _ in range(times):
            print(message)
#Example usage
if __name__ == "__main__":
    printer = Printer()

    printer.print_message("Hi")   #Default to 1 time
    printer.print_message("Hello", 4)
```
