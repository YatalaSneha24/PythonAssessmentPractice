```python
#Creating function and Implementation arithmetic operations using Python function

#Task 1:  Implementation of  arithmetic operations using Python function

#Function to add two numbers
def addition(y,s):
  result = y+s
  return result

#function to subtract two numbers
def subtraction(y,s):
  result = y-s
  return result

#Function to multiply two numbers
def multiplication(y,s):
  result = y*s
  return result

#Funtion to divide two numbers
def divide(y,s):
  if s != 0:
    result = y/s
  else:
    result = "Cannot divide by zero"
  return result

#Main program
number1 =int(input("Enter the first value: "))
number2 =int(input("Enter the second value: "))

#Calling the functions and printing the results
print("Addition: ", addition(number1,number2))
print("Subtraction: ", subtraction(number1, number2))
print("Multiplication: ", multiplication(number1, number2))
print("Division: ", divide(number1, number2))
```
