```python
#Implementation of End-to-End Set Operations

#Task 1: Implementation of Creating Sets
set_a = {1,2,3,4,5}
set_b = {4,5,6,7,8}
print("Set A:", set_a)
print("Set B:", set_b)

#Task 2: Implementation of Adding and Removing Elements in a Set
set_a.add(6)
set_a.remove(2)
print("Set A aftetr adding 6 and removing 2:", set_a)
print("Set A:", set_a)
print("Set B:", set_b)

# Task 3: Implementation of Union of Sets
union_set = set_a.union(set_b)
print("Union of Set A and Set B :", union_set)

#Task 4: Implementation of Intersection of Sets
intersection_set = set_a.intersection(set_b)
print("Intersection of Set A and Set B:", intersection_set)

#Task 5: Implementation of Difference of Sets
difference_set = set_a.difference(set_b)
print("Difference of Set A and Set B:", difference_set)

#Task 6: Implementation of Symmetric Difference of Sets
sym_diff_set = set_a.symmetric_difference(set_b)
print("Symmetric Difference of Set A and Set B:", sym_diff_set)

#Task 7: Implementation of Subset and Superset Testing
set_c = {3, 5}
is_subset = set_c.issubset(set_a)
is_superset = set_a.issuperset(set_c)
print(f"Is set C a subset of set A? {is_subset}")
print(f"Is set A a superset of set C? {is_superset}")

#Task 8: Implementation of Set Membership Testing
is_member = 3 in set_a
print("Is 3 in Set A", is_member)

#Task 9: Implementation of Removing Duplicates Using Sets
number_list = [1,2,3,4,4,5,5,6,7]
unique_numbers = set(number_list)  #Removing duplicates
print("List of numbers after removing duplicates:", unique_numbers)

#Task 10: Implementation of Set Comprehension
squares_set = {x** 2 for x in range(1,11)}
print("Set of squares of numbers from 1 to 10:", squares_set)

#Task 11: Implementation of Clearing a Set
set_a.clear()
print("Set A after clearing:", set_a)
```
