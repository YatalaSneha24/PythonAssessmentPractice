```python
#Implementation of Factorial Calculation and Fibonacci Sequence  using python recursion function

#Implementation of Factorial Calculation and Fibonacci Sequence  using python recursion function
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

n = int(input("Enter n:"))
print("Factorial of 5:", factorial(5))


```

```python
#fibanocci Sequence
def fibanocci(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibanocci(n-1) + fibanocci(n-2)

n = int(input("Enter n :"))
print("Fibanocci number at position", n, ":", fibanocci(n))
```
