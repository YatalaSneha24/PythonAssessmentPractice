```python
#Implementation of variable scope is crucial for managing variables effectively in functions

#Task 1:  Implementation of  Local Scope
def local_scope_ex():
  x=15 #Local variable
  print("Local variable x:", x)

local_scope_ex()
#print(x)
#This will raise a NameError because x is not accessible outside the function
```

```python
#Task 2:  Implementation of  Enclosing Scope (Nested Functions)
def outer_fun():
  x=10 #Enclosing variable
  def inner_fun():
    print("Enclosing variable x from inner function:", x)
  inner_fun()
outer_fun()
```

```python
#Task 3:  Implementation of  Global Scope

x = 15

def global_scope_ex():
  print("Global variable x:", x)

global_scope_ex()

print("Global variable x outside function:", x)
```

```python
#Task 4:  Implementation of  Modifying Global Variables

x = 2

def modify_global():
  global x
  x = 20  #modify global variable

modify_global()
print("Global variable x after modification:", x)
```
