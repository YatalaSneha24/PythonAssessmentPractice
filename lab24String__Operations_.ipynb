```python
#Implementation of End-to-End String Operations

#Task 1: Implementation of Creating and Accessing Strings
mi_string = "Hello, Sneha!"
print("Original String:", mi_string)
print("First character:", mi_string[0])
print("Last charcter:", mi_string[-1])

#Task 2: Implementation of String Concatenation and Repetition
str1 = "Yatala"
str2 = "Sneha"
conc_str = str1 + ' ' + str2
rep_str = conc_str * 4
print("Concatenated string:", conc_str)
print("Repetition string:",rep_str)

#Task 3: Implementation of String Case Manipulation
up_str = mi_string.upper()
lo_str = mi_string.lower()
ti_str = mi_string.title()
swa_str = mi_string.swapcase()
print("Uppercase:", up_str)
print("Lowercase:", lo_str)
print("Title case:", ti_str)
print("Swap case:", swa_str)

#Task 4: Implementation of Searching in Strings
substring = "Sneha"
fou_index = mi_string.find(substring)
if fou_index != -1:
  print(f"Substring '{substring}' found at index {fou_index}")
else:
  print(f"Substring '{substring}' not found.")

#Task 5: Implementation of Replacing Substrings
new_str = mi_string.replace("Sneha", "Pranavi")
print("String after replacement:", new_str)
new1_str = conc_str.replace("Sneha", "Pranavi")
print(new1_str)

#Task 6: Implementation of String Formatting
name = 'Anirudhuuu'
age = 18
for_str = f"My name is {name} and I am {age} years old."
print("Formatted string:", for_str)

#Task 7: Implementation of  Trimming and Padding String
text = "  extra spaces  "

#padding example (manually adding * characters)
padded = "*** Hello, Sneha!***"
print("Padded String:", padded)

#Trimming example (removing leading and trailing spaces)
trimmed = text.strip()
print("trimmed string:", trimmed)

#Task 8: Implementation of  Splitting and Joining Strings
sent = "Super Star Sneha papa"
words = sent.split()
print("Splitted words:", words)
join_sent = '-'.join(words)
print("Joined Sentence:", join_sent)

#Task 9: Implementation of  Counting Characters
char_count = mi_string.count('l')
print(f"Character 'l' appears {char_count} times in the string")

#Task 10: Implementation of  Checking String Properties
is_alpha = "SNeha".isalpha()
is_digit = "Yatala2".isdigit()
is_alnum = "HEllo24".isalnum()
is_space = "  ".isspace()

print("Is 'SNeha' alphabetic? ", is_alpha)
print("Is 'Yatala2?' numeric", is_digit)
print("Is 'HEllo24' alphanumeric?", is_alnum)
print("Is '  ' all spaces?", is_space)
```
