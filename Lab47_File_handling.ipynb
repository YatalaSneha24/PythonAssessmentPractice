```python
from os import write
#Implementation of the Basic Concepts of File Handling in Python

#Task 1: To Opening a file in write mode ('w')
file = open('example.ipynb', 'w')

file.write("Hello, Sneha!\nThis is a test file.")
file.close()

#Task 2: Opening a file in read mode ('r')
file = open('example.ipynb', 'r')
content = file.read()
print("Full Content: ")
print(content)

file.close()

#Task 3: Opening the file again  to read line by line
file = open('example.ipynb', 'r')
print("Reading Line by Line: ")
lines = file.readlines()
for line in lines:
  print(line, end='')

file.close()

#Task 4: Opening the file again to read all lines at once
file = open('example.ipynb', 'r')
print("\nReading All Lines at Once: ")
lines = file.readlines()
for line in lines:
  print(line, end='')

file.close()

#Task 5: Opening a file in append mode ('a')
file = open('example.ipynb', 'a')
file.write("\nAppending a new line to the file.")
file.close()

#Task 6: Using with statement to automatically close the file
with open('example.ipynb', 'r') as file:
  content = file.read()
  print("Content using 'with' Statement: ")
  print(content)
```
